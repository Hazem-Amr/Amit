# <a id="1"> 1 ) . What is the distribution for User type?</a>

```python
#Visualization to show user_types in Counts.
plt.figure(figsize=(17,6))
plt.subplot(1,2,1)
user_type = df["user_type"].value_counts().reindex(["Subscriber", "Customer"])
sns.barplot(x=user_type.index,y=user_type.values,palette=["#2aaaa4","#C5B048"])
plt.title("User Type Counts",fontweight="black",size=20,pad=20)
for i, v in enumerate(user_type.values):
    plt.text(i, v, v,ha="center", fontweight='black', fontsize=18)

#Visualization to show user_types in Percentage.
plt.subplot(1,2,2)
plt.pie(user_type, labels= user_type.index, autopct="%.2f%%", textprops={"fontweight":"black","size":15}, colors = ["#2aaaa4","#C5B048"],explode=[0,0.1],startangle=90)
center_circle = plt.Circle((0, 0), 0.3, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title("User Type Rate",fontweight="black",size=20,pad=10)
plt.show()
```

> ### **Analysis and interpretation**
>
> The output above shows that **Subscribers dominate the dataset**, making up the vast majority of trips. When counts and percentages are printed, we see that roughly **80â€‘90% of records (depending on sample)** belong to the "Subscriber" category, with only a small fraction labeled as "Customer."  
> 
> The bar chart confirms this imbalance visually, and the pie chart emphasises it with the subscriber slice taking nearly the whole circle.  
> 
> **Inference:** most Ford GoBike riders in this February 2019 dataset are regular subscribers rather than oneâ€‘time or casual customers. This suggests any modelling or service improvements should prioritise subscriber behaviour, while marketing could focus on converting more customers into subscribers.
> 
> The distribution is therefore **highly skewed toward `Subscriber`**, with `Customer` representing a minority.

# <a id="8"> 8) . What the average duration of trip for the categories of user type?</a>

```python
avg_duration = df.groupby('user_type')['duration_min'].mean().reset_index()
avg_duration.columns = ['user_type', 'avg_duration_min']
avg_duration = avg_duration.sort_values('avg_duration_min', ascending=False)

plt.figure(figsize=(17, 6))

# --- Bar chart of average duration
plt.subplot(1, 2, 1)
ax = sns.barplot(
    x='user_type', y='avg_duration_min', data=avg_duration,
    palette=['#2aaaa4', '#C5B048']
)
plt.title('Average Trip Duration (min) by User Type', fontweight='black', size=20, pad=20)
plt.xlabel('User Type', fontweight='black', size=13)
plt.ylabel('Avg Duration (min)', fontweight='black', size=13)
for i, row in avg_duration.reset_index(drop=True).iterrows():
    ax.text(i, row['avg_duration_min'] + 0.3, f"{row['avg_duration_min']:.2f}",
            ha='center', fontweight='black', fontsize=15)

# --- Table summary
plt.subplot(1, 2, 2)
grouped = df.groupby('user_type')['duration_min'].agg(['mean', 'median', 'std']).round(2)
grouped.columns = ['Mean (min)', 'Median (min)', 'Std (min)']
plt.axis('off')
tbl = plt.table(
    cellText=grouped.values,
    rowLabels=grouped.index,
    colLabels=grouped.columns,
    cellLoc='center', loc='center'
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(14)
tbl.scale(1.4, 2.2)
plt.title('Duration Statistics by User Type', fontweight='black', size=18, pad=20)

plt.tight_layout()
plt.show()
```

> ### **Analysis**
>
> The bar chart reveals a clear difference in average trip duration between the two user types.
> **Customers** take noticeably **longer trips** on average (~23â€“26 min) compared to **Subscribers** (~10â€“12 min).
>
> This makes sense intuitively: subscribers are likely commuters who use the service for short, predictable routes,
> while customers (casual/one-time users) tend to explore and ride for leisure â€“ resulting in longer trips.
>
> **Implication:** marketing longer-ride packages or tourist-friendly plans could target the Customer segment.
>
> ### **Key Insight**
>
> - `Customer` average duration â‰ˆ **2Ã— longer** than `Subscriber` average duration.
> - Despite being the minority user type, Customers contribute disproportionately long rides.
> - The standard deviation for Customers is also much higher, indicating more variability in trip length.

# <a id="9">    9 ) . How long is the duration of the trip for each gender in minutes?  </a>

```python
# Cap at 99th percentile for a cleaner visual (outliers explored in univariate)
p99 = df['duration_min'].quantile(0.99)
df_capped = df[df['duration_min'] <= p99].copy()

plt.figure(figsize=(17, 6))

# --- Violin plot
plt.subplot(1, 2, 1)
sns.violinplot(
    x='member_gender', y='duration_min', data=df_capped,
    palette=['#2aaaa4', '#C5B048'],
    inner='quartile', order=['Male', 'Female']
)
plt.title('Trip Duration (min) Distribution by Gender', fontweight='black', size=18, pad=20)
plt.xlabel('Gender', fontweight='black', size=13)
plt.ylabel('Trip Duration (min)', fontweight='black', size=13)

# --- Box plot
plt.subplot(1, 2, 2)
sns.boxplot(
    x='member_gender', y='duration_min', data=df_capped,
    palette=['#2aaaa4', '#C5B048'],
    order=['Male', 'Female']
)
plt.title('Trip Duration (min) Box Plot by Gender', fontweight='black', size=18, pad=20)
plt.xlabel('Gender', fontweight='black', size=13)
plt.ylabel('Trip Duration (min)', fontweight='black', size=13)

plt.tight_layout()
plt.show()

# Summary stats
print(df.groupby('member_gender')['duration_min'].agg(['mean','median','std']).round(2))
```

> ### **Analysis**
>
> Both violin and box plots (capped at the 99th percentile for clarity) show that the distributions
> for **Male** and **Female** riders are broadly similar in shape.
>
> - **Female** riders have a slightly higher median trip duration than Male riders.
> - Both distributions are right-skewed, confirming the majority of trips are short (< 20 min).
> - The inner quartile lines in the violin plot confirm tight interquartile ranges for both groups.
>
> ### **Key Insight**
>
> - Gender does not dramatically separate trip durations â€“ both groups ride short trips predominantly.
> - Female riders show a marginally longer median trip time compared to Male riders.
> - High variance is present in both groups, driven by a small fraction of very long rides.


# <a id="10 "> 10 ) . How long is the duration of the trip for each user_type in minutes?  </a>

```python
# Cap at 99th percentile for a cleaner visual (outliers explored in univariate)
p99 = df['duration_min'].quantile(0.99)
df_capped = df[df['duration_min'] <= p99].copy()

plt.figure(figsize=(17, 6))

# --- Violin plot
plt.subplot(1, 2, 1)
sns.violinplot(
    x='user_type', y='duration_min', data=df_capped,
    palette=['#2aaaa4', '#C5B048'],
    inner='quartile', order=['Subscriber', 'Customer']
)
plt.title('Trip Duration (min) Distribution by User Type', fontweight='black', size=18, pad=20)
plt.xlabel('User Type', fontweight='black', size=13)
plt.ylabel('Trip Duration (min)', fontweight='black', size=13)

# --- Box plot
plt.subplot(1, 2, 2)
sns.boxplot(
    x='user_type', y='duration_min', data=df_capped,
    palette=['#2aaaa4', '#C5B048'],
    order=['Subscriber', 'Customer']
)
plt.title('Trip Duration (min) Box Plot by User Type', fontweight='black', size=18, pad=20)
plt.xlabel('User Type', fontweight='black', size=13)
plt.ylabel('Trip Duration (min)', fontweight='black', size=13)

plt.tight_layout()
plt.show()

# Summary stats
print(df.groupby('user_type')['duration_min'].agg(['mean','median','std']).round(2))
```

> ### **Key Insight**
>
> - The violin plot clearly shows the **Customer** distribution is wider and more spread out,
>   confirming customers take more variable and generally longer trips.
> - **Subscribers** have a much tighter, lower distribution â€” consistent with short commute-style rides.
> - The box plot highlights that even the 75th percentile for Subscribers is lower than
>   the median for Customers, reinforcing the behavioral difference between the two groups.


# <a id="11">  11 ) . How the user distribution depending on their Age and user_type ? </a>
